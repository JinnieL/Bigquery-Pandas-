# 2. Data Mart

데이터 분석과 데이터 마트의 연관성    
- 데이터 마트는 데이터를 구조화, 통합하여 분석을 더 쉽고 빠르게 할 수 있도록 도와줌.
- 다양한 소스의 데이터를 효율적으로 저장하고 필요한 정보를 빠르게 추출하여 비즈니스 인사이트 도출에 도움.

## 1. 데이터 마트

### 1-1. 데이터 마트란

- 필요한 데이터를 확인할 수 있게 분류된 데이터가 저장된 저장소

### 1-2. 데이터 마트 특징
- 데이터 가용성이 높고 접근성에서 편리
    - 데이터 가용성 : 데이터를 적절하게 이용 가능한 정도.
- 효율적인 데이터 관리가 가능.

사용자의 업무에 따라 데이터 마트를 구성하기 때문에 쉽게 접근 및 활용 가능.

## 2. 데이터 마트 생성방법 + ETL

스크린샷 2024-09-19 오후 5.44.08.png

### ETL : 데이터 추출 프로세스
Extraction : 데이터 수집     
Transformation : 추출된 데이터를 일관된 형식으로 변환하여 효율적으로 데이터 관리 및 검색 기능을 제공.       
Load : 추출, 변환 완료된 데이터를 저장소(DB)에 입력 및 저장.        
저장된 데이터는 하나의 데이터마트로 구성되어 대시보드 연결 및 레포트 생성 등 가능하게 됨.


## 3. 데이터 분석에 ETL이 필요한 이유

다른 형식의 데이터를 한꺼번에 보는 등의 경우, 데이터 확인 및 검색의 효율성을 위해 데이터를 변환하여 저장.

## 4. ETL의 활용 - 대시보드

스크린샷 2024-09-19 오후 5.55.37.png

데이터마트가 구성되어 있지 않다면 데이터를 일일히 전처리한 뒤에 쿼리문을 작성하므로 효율적이지 않으며, 매번 수정해야 하는 부담이 있음.      
반면, 데이터마트가 구성되어 있다면 쿼리문 작성만으로도 필요한 데이터를 바로 대시보드로 전달 할 수 있기에 속도 및 비용 측면에서 매우 효율적이다.
